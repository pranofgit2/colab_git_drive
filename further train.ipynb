```python
import keras 
import numpy as np 
import pickle
```

```python
from google.colab import drive
from os.path import join
```

```python
ROOT = '/content/drive'
PROJ = 'My Drive/Colab Notebooks/MLOps'

drive.mount(ROOT)
PROJECT_PATH = join(ROOT, PROJ)
```

```python
X_train=np.load(f'{PROJECT_PATH}/X_train.npy')
X_test=np.load(f'{PROJECT_PATH}/X_test.npy')


y_train=np.load(f'{PROJECT_PATH}/y_train.npy')
y_test=np.load(f'{PROJECT_PATH}/y_test.npy')
```

```python
GIT_REPO_NAME='colab_git_drive/'
REPO_PATH=join(PROJECT_PATH,GIT_REPO_NAME)
MODEL_NAME='md.h5'
```

```python
# load model
model=keras.models.load_model(f'{REPO_PATH}/{MODEL_NAME}')
```

```python
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=20,epochs=2,verbose=1)
```

```python
# saving model history

with open(f'{REPO_PATH}/hist_dict', 'wb') as file_pi:
    pickle.dump(history.history, file_pi)


# saving model 
model.save(f'{REPO_PATH}/{MODEL_NAME}')
```
